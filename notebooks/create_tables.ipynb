```python
#import pandas and sqlite
import pandas as pd
import sqlite3 as sl#connecct to database object
conn = sl.connect('Elections.db')
c = conn.cursor()
```

```python
c.execute('DROP TABLE IF EXISTS fivethirtyeight')
c.execute('CREATE TABLE fivethirtyeight (cycle str, branch str, type str, matchup str, forecastdate datetime, state str, startdate datetime, enddate datetime, pollster str, grade str, samplesize str, population str, poll_wt str, rawpoll_clinton, rawpoll_trump, rawpoll_johnson, rawpoll_mcmullin, adjpoll_clinton, adjpoll_trump, adjpoll_johnson, adjpoll_mcmullin, multiversions, url, poll_id, question_id, createdate datetime, timestamp datetime)')
conn.commit()

fivethirtyeight.to_sql('fivethirtyeight',conn, if_exists='replace')
```

```python
df = pd.read_csv('https://projects.fivethirtyeight.com/polls-page/president_polls.csv')
df.columns
```

```python
c.execute('DROP TABLE IF EXISTS trumpbiden')
c.execute('CREATE TABLE trumpbiden (question_id, poll_id, cycle, state, pollster_id, pollster, sponsor_ids, sponsors, display_name, pollster_rating_id, pollster_rating_name, fte_grade, sample_size, population, population_full, methodology, office_type, seat_number, seat_name, start_date, end_date, election_date, sponsor_candidate, internal, partisan, tracking, nationwide_batch, ranked_choice_reallocated, created_at, notes, url, stage, answer, candidate_name, candidate_party, pct)')
conn.commit()

df.to_sql('trumpbiden',conn, if_exists = 'replace')
```
