```python
# importing some packages to explore the data
import numpy as np
import pandas as pd

pd.set_option("display.max_columns", 50)
pd.set_option("display.max_rows", 50)
```

```python
clinton_url = "http://projects.fivethirtyeight.com/general-model/president_general_polls_2016.csv"
biden_url = "https://projects.fivethirtyeight.com/polls-page/president_polls.csv"

clinton = pd.read_csv(clinton_url)
biden = pd.read_csv(biden_url)

biden.head()
```

```python
biden['state'] = biden.state.fillna('US')
```

```python
biden
```

```python
def opponent (biden, opp):
    trump_vs = biden[(biden['answer'] == opp) | (biden['answer'] == 'Trump')]
    trump_vs = trump_vs.pivot_table(values = "pct", index = ["poll_id", "start_date", "end_date", "state", "pollster", "fte_grade", "methodology", "sample_size", "population"], columns = "answer")
    trump_vs = trump_vs.dropna(axis = 0, how = 'any') #Drops the Trump polls against any opponent that isn't our opp parameter
    trump_vs = trump_vs.reset_index().drop(columns = ["poll_id"])
    trump_vs["start_date"] = pd.to_datetime(trump_vs["start_date"])
    trump_vs["end_date"] = pd.to_datetime(trump_vs["end_date"]) 
    trump_vs["dem_lead"] = trump_vs[opp] - trump_vs["Trump"] 
    trump_vs = trump_vs.sort_values(by = ["end_date", "start_date"]) #Arranging the polls from most to least recent
    return trump_vs
```

```python
trump_biden = opponent(biden, 'Biden')
trump_biden.head()
```

```python
mask = trump_biden['state'] == 'US'
```

```python
biden_us = trump_biden[mask]
biden_us
```
